```python
from sklearn.feature_selection import VarianceThreshold
```

```python
df = pd.read_pickle("gene_pav_data.pkl")
```

```python
df.head()
```

```python
selector = VarianceThreshold(threshold=0.01)
```

```python
df_high_variance = selector.fit_transform(df)
```

```python
df_high_variance.shape
```

```python
selected_features_mask = selector.get_support()
```

```python
selected_feature_names = df.columns[selected_features_mask]
```

```python
df_selected = pd.DataFrame(df_high_variance, columns=selected_feature_names, index=df.index)
```

```python
df_selected.head()
```

```python
df_selected.info()
```

```python
df_selected.to_pickle("gene_pav_afterFS_varthreshold.pkl")
```
