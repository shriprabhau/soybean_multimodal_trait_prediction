```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
import pickle
```

```python
pheno = pd.read_csv ('yield_pheno.txt', sep = '\t')
```

```python
pheno.head()
```

```python
pheno.info()
```

```python
pheno['new_line'] = pheno['Data-storage-ID'] + '_' + pheno['method_name']
```

```python
pheno.head()
```

```python
pheno.info()
```

```python
# Extract the specified column
column_data = pheno[['method_name']]

# Initialize OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)  # Drop first category to avoid multicollinearity

# Fit and transform the OneHotEncoder
one_hot_encoded = ohe.fit_transform(column_data)
```

```python
# Get the column names after one-hot encoding
categories = ohe.get_feature_names_out(['method_name'])
#the category name i as followed - Location_year_BeiJing_2013. 
#So split based on underscores and pull the 2-4 parts to used as column names

feature_names = ['_'.join(col.split('_')[2:5]) for col in categories]
```

```python
# Create a DataFrame with the one-hot encoded columns
one_hot_df = pd.DataFrame(one_hot_encoded, columns=feature_names)
```

```python
#Check if ohe is done right
one_hot_df.head()
```

```python
df = one_hot_df.astype(int)

print(df)
```

```python
# Concatenate the one-hot encoded columns to the original DataFrame
ohe_data = pd.concat([pheno.reset_index(drop=True), df], axis=1)
ohe_data.head()
```

```python
# Drop the original 'Location_Year' column
ohe_data = ohe_data.drop(['method_name'], axis=1)
```

```python
# Reorganize the columns to make 'PRT' the last column
column_order = [col for col in ohe_data.columns if col != 'observation_value'] + ['observation_value']
ohe_data = ohe_data[column_order]


# Display the result
ohe_data.head()
```

```python
ohe_data.to_pickle("phenotype_yield_data.pkl")
```
