```python
#Load pfam
pfam = pd.read_csv('../interproscan/gene_pfam_eval5.txt', sep="\t", header=None)
```

```python
pfam.columns = pfam.iloc[0]  # set first row as columns
pfam = pfam[1:]   
pfam = pfam.set_index("Protein_ID")
```

```python
pfam.head()
```

```python
def extract_gene_id(protein_id):
    # match patterns like GlymaLee.U002300.1.p or UWASoyPan00005.t1
    m = re.search(r'([A-Za-z]*[U0-9]+[A-Za-z0-9]*)', protein_id)
    return m.group(1) if m else protein_id

    #its regex pattern search tries to find something that looks like a compact gene or protein ID, possibly starting with letters, followed by digits, followed by more alphanumeric characters
    #The first is telling first character could be letters or numbers or U and the string after '+' is telling it could be letters or numbers
    #giving U separately as I know some proteins start way
```

```python
pfam["Protein_part"] = pfam.index.to_series().apply(extract_gene_id)
```

```python
pfam["Pfam_protein"] = pfam["Pfam_domain"] + "_" + pfam["Protein_part"]
pfam = pfam.drop(columns=["Pfam_domain", "Protein_part"])
```
